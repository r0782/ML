```python
import numpy as np

# Create array
arr = np.array([1,2,3,4,5,6])

# Slicing
print("Slicing:")
print(arr[1:5])

# Reshaping
reshaped = arr.reshape(2,3)
print("\nReshaped Array:")
print(reshaped)

# Broadcasting
broadcasted = reshaped + 10
print("\nBroadcasting:")
print(broadcasted)

```
